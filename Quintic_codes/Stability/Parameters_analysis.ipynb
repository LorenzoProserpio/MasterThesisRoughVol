```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
param = np.array(pd.read_csv("quintic_stability_parameters.csv"))
dates = np.array(["23/01", "24/01", "25/01", "26/01", "27/01", "30/01", "06/02", "13/02", "21/02"])
title = np.array(["rho", "H", "epsilon", "alpha_0", "alpha_1", "alpha_3", "alpha_5"])
```

```python
std = param.std(axis = 0)
var = param.var(axis = 0)

for i in range(7):
    print(f"{title[i]}\t\tStandard deviation: {std[i]: .5f}\tVariance: {var[i]: .5f}\n")
```

```python
var[5]
```

```python
fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(10)
plt.plot(title, std)
plt.title("Standard Deviation")
```

```python
fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(10)
plt.plot(title, var)
plt.title("Variance")
```

```python
title = np.array([r"$\rho$", "H", r"$\varepsilon$", r"$\alpha_0$", r"$\alpha_1$", r"$\alpha_3$", r"$\alpha_5$"])

for i in range(7):
    fig = plt.figure(i)
    fig.set_figheight(5)
    fig.set_figwidth(10)
    plt.plot(dates, param[:,i])
    plt.title(title[i])
```
